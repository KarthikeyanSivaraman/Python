```python
# Create a function that checks whether the given number is Odd or Even
class OddEven():
    def OddEven():
        number = int(input("Enter a number: "))
        if number % 2 == 0:
            text = 'is Even number'
        else:
            text = 'is Odd number'
        return str(number) + " " + text
print(OddEven.OddEven())
```

```python
# Create a function that tells elegibility of marriage for male and female according to their age limit like 21 for male and 18 for female
class EligibilityForMarriage():
    def eligible():
        gender = input("Enter your gender (Male/Female): ")
        age = int(input("Enter your age: "))
        if (gender == 'Male' and age >= 21) or (gender == 'Female' and age >= 18):
            print("ELIGIBLE")
        else:
            print("NOT ELIGIBLE")

# Call the function
EligibilityForMarriage.eligible()
```

```python
# calculate the percentage of your 10th marks
class FindPercent():
    def percentage():
        Subject1 = int(input('Enter your Subject1 marks: '))
        Subject2 = int(input('Enter your Subject2 marks: '))
        Subject3 = int(input('Enter your Subject3 marks: '))
        Subject4 = int(input('Enter your Subject4 marks: '))
        Subject5 = int(input('Enter your Subject5 marks: '))
        Total = Subject1 + Subject2 + Subject3 + Subject4 + Subject5
        Percentage = (Total / 500) * 100    
        return Percentage
print(FindPercent.percentage()) 
```
